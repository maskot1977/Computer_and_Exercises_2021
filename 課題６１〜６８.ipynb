```python
def numerical_differentiation(f, x, h=1e-4):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
def gradient_decent(f, x0, lr=1e-4, step_num=200, h=1e-4):
    x = x0
    for _ in range(step_num):
        grad = numerical_differentiation(f, x, h=h)
        x = x - lr * grad
    return x
```

```python
def gradient_decent(f, x0, lr=1e-4, step_num=200, h=1e-4):
    grad_history = []
    x_history = []
    x = x0
    for _ in range(step_num):
        grad = numerical_differentiation(f, x, h=h)
        grad_history.append(grad)
        x_history.append(x)
        x = x - lr * grad
    return x_history, grad_history
```

```python
x_history, grad_history = gradient_decent(f, 4, lr=1e-2, step_num=300, h=1e-4)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_history)
plt.grid()
```

```python
import matplotlib.pyplot as plt
plt.plot(grad_history)
plt.grid()
```

```python
import matplotlib.pyplot as plt

def colormap(f, x_min, x_max, y_min, y_max, h=0.1):
    x_mg, y_mg = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = [[f([x, y]) for x, y in zip(xx, yy)] for xx, yy in zip(x_mg, y_mg)]
    plt.contourf(x_mg, y_mg, Z, cmap='bwr') #plt.cm.RdBu)
    plt.colorbar()
    plt.grid()
    plt.show()
```

```python
def surfaceplot(f, x_min, x_max, y_min, y_max, h=0.1):
    x_mg, y_mg = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = [[f([x, y]) for x, y in zip(xx, yy)] for xx, yy in zip(x_mg, y_mg)]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x_mg, y_mg, np.array(Z), cmap='bwr', linewidth=0)
    fig.colorbar(surf)
    ax.set_title("Surface Plot")
    fig.show()
```

```python
surfaceplot(f, -20, 20, -20, 20, h=0.1)
```

```python
surfaceplot(g, -20, 20, -20, 20, h=0.1)
```

```python
def numerical_differentiation(f, x, h=1e-4):
    grad = np.zeros_like(x)
    for i in range(len(x)):
        tmp = x[i]

        # f(x + h)
        x[i] = tmp + h
        fxh1 = f(x)

        # f(x - h)
        x[i] = tmp - h
        fxh2 = f(x)

        grad[i] = (fxh1 - fxh2) / (2 * h) 
        x[i] = tmp       

    return grad
```

```python
x_history, grad_history = gradient_decent(f, [10, 10], lr=1e-2, step_num=300, h=1e-4)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_history)
plt.grid()
```

```python
import matplotlib.pyplot as plt
plt.plot(grad_history)
plt.grid()
```

# 課題６７

```python
import scipy as sp # 科学計算ライブラリ
from scipy import stats # 統計計算ライブラリ

significance = 0.05
o = [17, 10, 6, 7, 15, 5] # 分布１（ex.実測値）
e = [10, 10, 10, 10, 10, 10] # 分布２（ex.理論値）

chi2, p = stats.chisquare(o, f_exp = e)

print('chi2 値は %(chi2)s' %locals())
print('確率は %(p)s' %locals())

if p < significance:
    print('有意水準 %(significance)s で、有意な差があります' %locals())
else:
    print('有意水準 %(significance)s で、有意な差がありません' %locals())
```
