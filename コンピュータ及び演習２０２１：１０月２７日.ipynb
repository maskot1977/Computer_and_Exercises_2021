# 課題３３

```
def question33(x):
    V = [500, 100, 50, 10, 5, 1]
    C = [10, 10, 10, 10, 10, 10]

    nanmai = 0
    for v, c in zip(V, C):
        t = min(int(x / v), c) 
        x -= t * v
        print(v, "円硬貨", t, "枚")
        nanmai += t
        if t < 0:
            return False

    print("残額", x)
    if x > 0:
        return False

    return nanmai
```

```
x =  1672
question33(x)
```

```
x =  8802
question33(x)
```

```
x =  -124
question33(x)
```

# 課題３６

```
def question36(n):
    result = []
    for i in range(2, round(math.sqrt(n))):
        while n%i == 0:
            result.append(i)
            n /= i
    if n != 1:
        result.append(int(n))
    return sorted(result)
```

# 課題３７

```
def question37(n):
    result = []
    is_prime = [True for _ in range(n + 1)]
    is_prime[0] = False
    is_prime[1] = False
    
    for i in range(2, n + 1):
        #print(is_prime)
        #if is_prime[i] == True:
        if is_prime[i]:
            result.append(i)
            j = 2 * i
            while j <= n:
                is_prime[j] = False
                j += i
    
    return len(result)
```

# 課題３８

```
def gcd(a, b): # ユークリッドの互除法
    if b == 0:
        return a
    return gcd(b, a%b)
```

```
def question38(P, Q):
    return gcd(abs(P[0] - Q[0]), abs(P[1] - Q[1])) - 1
```

```
x1 =  -2
y1 =  -9
x2 =  6
y2 =  7
question38([x1, y1], [x2, y2])
```

```
%matplotlib inline
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(3,6))
ax = fig.add_subplot(111)
ax.plot([x1, x2], [y1, y2])
ax.set_xticks(range(x1, x2+1, 1))
ax.set_yticks(range(y1, y2+1, 1))
plt.grid()
```

# 課題３９

```
import math
def question39(P, t):
    rad = t * math.pi / 180
    a = [
         [math.cos(rad), -math.sin(rad)],
         [math.sin(rad), math.cos(rad)]
    ]
    return [
            P[0] * a[0][0] + P[1] * a[0][1],
            P[0] * a[1][0] + P[1] * a[1][1]
            ]
```

```
P = [1, 0]
t = 45
question39(P, t)
```

```
P = [0, 1]
t = 30
question39(P, t)
```

```
import numpy as np
def question39_numpy(P, t):
    rad = t * np.pi / 180
    a = np.array([
         [np.cos(rad), -np.sin(rad)],
         [np.sin(rad), np.cos(rad)]]
    )
    return np.dot(a, P)
```

```
P = [1, 0]
t = 45
question39_numpy(P, t)
```

```
P = [0, 1]
t = 30
question39_numpy(P, t)
```

# 課題４０

```
import math
def question39(P, t):
    rad = t * math.pi / 180
    a = [
         [math.cos(rad), -math.sin(rad)],
         [math.sin(rad), math.cos(rad)]
    ]
    return [
            P[0] * a[0][0] + P[1] * a[0][1],
            P[0] * a[1][0] + P[1] * a[1][1]
            ]
```

```
def question40(P, n):
    angle = 360 / n
    Q = []
    Q.append(P)
    for i in range(n):
        x = question39(Q[i], angle)
        Q.append(x)
    return Q
```

```
%matplotlib inline
import matplotlib.pyplot as plt

P = [0, 1]
n = 5
Q = question40(P, n)
plt.figure(figsize=(6,6))
plt.plot([q[0] for q in Q], [q[1] for q in Q])
plt.show()
```

# 課題４１

```
import math
def question39(P, t):
    rad = t * math.pi / 180
    a = [
         [math.cos(rad), -math.sin(rad)],
         [math.sin(rad), math.cos(rad)]
    ]
    return [
            P[0] * a[0][0] + P[1] * a[0][1],
            P[0] * a[1][0] + P[1] * a[1][1]
            ]
```

```
def question41(P, n):
    angle = 360 / n
    Q = []
    Q.append(P)
    for i in range(n):
        x = question39(Q[i], angle * 2)
        Q.append(x)
    return Q
```

```
%matplotlib inline
import matplotlib.pyplot as plt

P = [0, 1]
n = 5
Q = question41(P, n)
plt.figure(figsize=(6,6))
plt.plot([q[0] for q in Q], [q[1] for q in Q])
plt.show()
```

```
def question41b(P, n):
    angle = 360 / n
    Q = []
    Q.append(P)
    for i in range(n):
        x = question39(Q[i], angle * int(n/2))
        Q.append(x)
    return Q
```

```
%matplotlib inline
import matplotlib.pyplot as plt

P = [0, 1]
n = 19
Q = question41b(P, n)
plt.figure(figsize=(6,6))
plt.plot([q[0] for q in Q], [q[1] for q in Q])
plt.show()
```

```
def question41c(P, n, m):
    angle = 360 / n
    Q = []
    Q.append(P)
    for i in range(n):
        x = question39(Q[i], angle * m)
        Q.append(x)
    return Q
```

```
%matplotlib inline
import matplotlib.pyplot as plt

P = [0, 1]
n = 19
Q = []
plt.figure(figsize=(6,6))
for m in [2, 3, 4, 5, 6, 7, 8, 9, 10]:
    Q += question41c(P, n, m)
plt.plot([q[0] for q in Q], [q[1] for q in Q])
plt.show()
```
