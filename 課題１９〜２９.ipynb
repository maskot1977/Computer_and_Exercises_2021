# 課題１９


```python
def fibonacci_recursive(n):
    if n <= 1:
        return n

    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
```

```python
%%timeit
fibonacci_recursive(10)
```

```python
def fibonacci_not_recursive(n):
    if n <= 1:
        return n

    else:
        a = 0
        b = 1
        for i in range(n - 1):
            c = a + b
            a = b
            b = c

        return b
```

```python
%%timeit
fibonacci_not_recursive(11)
```

```python
product_nm = ["Vanilla", "Strawberry", "Green tea", "Chocolate", "Mint"]
sales = [132, 60, 70, 120, 25]

def sort_values(product_nm, sales):
    data = []
    for x, y in zip(product_nm, sales):
        data.append([x, y])

    names = []
    values = []
    data = sorted(data, key=lambda x: x[1], reverse=True)
    for x, y in data:
        names.append(x)
        values.append(y)

    return names, values
```

```python
product_nm, sales = sort_values(product_nm, sales)
```

```python
plt.title("What was the best-selling flavors?", fontsize = 22) 
plt.rcParams['font.size'] = 16.0 

plt.pie(sales, labels=product_nm, counterclock=False, startangle=90,
        autopct="%1.1f%%", 
        colors=pie_colors,
        explode=[0, 0, 0, 0, 0], 
        shadow=True, 
        wedgeprops={'linewidth': 1, 'edgecolor':"black"}) 
plt.axes().set_aspect(1.0)
plt.show()
```

# 課題２６

```python
# 分散を求める関数を作ろう
def variance(list_):
    ave = mean(list_)
    sum = 0
    for x in list_:
        sum += (x - ave)**2
    return sum / len(list_)
```

```python
variance(col2)
```

```python
variance(col3)
```

```python
variance(col4)
```

```python
# 標準偏差を求める関数を作ろう
import math
def standard_deviation(list_):
    return math.sqrt(variance(list_))
```

# 課題２７

```python
# 共分散 = 偏差積の平均 （偏差値、ではありません。偏差積、です）を作ろう
def covariance(list1, list2): 
    sum = 0
    for x, y in zip(list1, list2):
        sum += (x - mean(list1)) * (y - mean(list2))

    return sum / len(list1)
```

```python
covariance(col2, col3)
```

```python
covariance(col3, col4)
```

```python
covariance(col4, col2)
```

```python
covariance(col2, col2)
```

# 課題２８

```python
# 相関係数 = 共分散を list1, list2 の標準偏差で割ったものを作ろう
def correlation(list1, list2):
    return covariance(list1, list2) / standard_deviation(list1) / standard_deviation(list2)
```

```python
for i, x in enumerate([col1, col2, col3, col4, col5, col6, col7]):
    for j, y in enumerate([col1, col2, col3, col4, col5, col6, col7]):
        if i >= j:
            continue

        print(i, j, correlation(x, y))
```

```python
for i, x in enumerate([col1, col2, col3, col4, col5, col6, col7]):
    for j, y in enumerate([col1, col2, col3, col4, col5, col6, col7]):
        if i >= j:
            continue

        plt.title(correlation(x, y))
        plt.xlabel(i)
        plt.ylabel(j)
        plt.scatter(x, y)
        plt.show()
```

# 課題２９

```python
# 回帰直線の傾き＝相関係数＊（（yの標準偏差）／（xの標準偏差））を求める関数を作ろう
def a_fit(xlist, ylist):
    return correlation(xlist, ylist) * standard_deviation(ylist) / standard_deviation(xlist)
```

```python
# y切片＝yの平均－（傾き＊xの平均）を求める関数を作ろう
def b_fit(xlist, ylist):
    return mean(ylist) - a_fit(xlist, ylist) * mean(xlist)
```

```python
# 回帰直線の式を表示
X = col3
Y = col2
print("y = f(x) = {0} x + {1}".format(a_fit(X, Y), b_fit(X, Y)))
```

```python
def f(x):
    return a_fit(X, Y) * x + b_fit(X, Y)
```
