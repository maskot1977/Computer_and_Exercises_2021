# 課題４１

```python
def task41(f, x_0=1, x_n=2, h=0.01):
    sum_ = 0
    n = int((x_n - x_0) / h)
    for i in range(1, n + 1):
        x_i = x_0 + h * i
        sum_ += f(x_i) * h
    return sum_
```

```python
def f(x):
    return x / (x**2 + 1)
```

```python
task41(f, x_0=1, x_n=2, h=0.01)
```

```python
task41(f)
```

```python
task41(f, h=1e-4)
```

```python
for z in range(1, 7):
    h = 10**(-z)
    print(z, h, task41(f, h=h))
```

```python
import matplotlib.pyplot as plt

hs41 = []
ys41 = []
for z in range(1, 8):
    h = 10**(-z)
    y = task41(f, h=h)
    hs41.append(h)
    ys41.append(y)

plt.plot(hs41, ys41, marker='o')
plt.xscale('log')
```

```python
import timeit

timeit.timeit(lambda: task41(f, h=1e-6), number=1)
```

```python
import matplotlib.pyplot as plt

hs41 = []
ys41 = []
ts41 = []
for z in range(1, 8):
    h = 10**(-z)
    y = task41(f, h=h)
    t = timeit.timeit(lambda: task41(f, h=h), number=1)
    hs41.append(h)
    ys41.append(y)
    ts41.append(t)

plt.plot(hs41, ys41, marker='o')
plt.xscale('log')
plt.show()
plt.plot(hs41, ts41, marker='o')
plt.xscale('log')
plt.show()
```

# 課題４２

```python
def task42(f, x_0=1, x_n=2, h=0.01):
    sum_ = 0
    n = int((x_n - x_0) / h)
    for i in range(1, n + 1):
        x_i = x_0 + h * i
        sum_ += (f(x_i) + f(x_i - h)) * h / 2
    return sum_
```

```python
def f(x):
    return x / (x**2 + 1)
```

```python
task42(f)
```

```python
import matplotlib.pyplot as plt

hs42 = []
ys42 = []
ts42 = []
for z in range(1, 8):
    h = 10**(-z)
    y = task42(f, h=h)
    t = timeit.timeit(lambda: task42(f, h=h), number=1)
    hs42.append(h)
    ys42.append(y)
    ts42.append(t)

plt.plot(hs42, ys42, marker='o')
plt.xscale('log')
plt.show()
plt.plot(hs42, ts42, marker='o')
plt.xscale('log')
plt.show()
```

# 課題４３

```python
def task43(f, x_0=1, x_n=2, h=0.001):
    sum_ = 0
    n = int((x_n - x_0) / (h * 2))
    for i in range(1, n + 1):
        x_i = x_0 + (2 * i - 1) * h
        sum_ += f(x_i + h) + 4 * f(x_i) + f(x_i - h)
    return sum_ * h / 3
```

```python
task43(f)
```

# 課題４４

```python
def task44(f, a, b, e=0.00001):
    if f(a) * f(b) > 0:
        return False

    while abs(f(a) - f(b)) > e:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) > 0:
            a = c
        else:
            b = c
    
    return (a + b) / 2
```

```python
import math
def f(x):
    return math.exp(x) - 3 * x
```

```python
task44(f, 0, 1)
```

```python
task44(f, 0, 1, e=0.01)
```

```python
import matplotlib.pyplot as plt
w = 100
X = [x * 0.01 for x in range(w)]
Y = [f(x * 0.01) for x in range(w)]
plt.plot(X, Y)
plt.grid()
```

```python
import timeit
import matplotlib.pyplot as plt

hs44 = []
ys44 = []
ts44 = []
for z in range(1, 8):
    e = 10**(-z)
    y = task44(f, 0, 1, e=e)
    t = timeit.timeit(lambda: task44(f, 0, 1, e=e), number=1)
    hs44.append(e)
    ys44.append(y)
    ts44.append(t)

plt.plot(hs44, ys44, marker='o')
plt.xscale('log')
plt.show()
plt.plot(hs44, ts44, marker='o')
plt.xscale('log')
plt.show()
```

# 課題４５

```python
import math
def task45(f, df, x, e=0.00001):
    if abs(f(x)) < e:
        return x
    else:
        x2 = x - f(x) / df(x)
        if max(abs(x2 - x), abs(f(x2) - f(x))) < e:
               return x2
        else:
            return task45(f, df, x2, e=e)
```

```python
def f(x):
    return math.exp(x) - 3 * x

def df(x):
    return math.exp(x) - 3
```

```python
task45(f, df, 0)
```

```python
import timeit
import matplotlib.pyplot as plt

hs45 = []
ys45 = []
ts45 = []
for z in range(1, 8):
    e = 10**(-z)
    y = task45(f, df, 0, e=e)
    t = timeit.timeit(lambda: task45(f, df, 0, e=e), number=1)
    hs45.append(e)
    ys45.append(y)
    ts45.append(t)

plt.plot(hs45, ys45, marker='o')
plt.xscale('log')
plt.show()
plt.plot(hs45, ts45, marker='o')
plt.xscale('log')
plt.show()
```

# 課題４６

```python
def distance(x, y):
    return abs(x - y)

def task46(H):
    dp_table = [0, distance(H[0], H[1])]
    for i in range(2, len(H)):
        dp_table.append(
            min(
                dp_table[i - 2] + distance(H[i - 2], H[i]),
                dp_table[i - 1] + distance(H[i - 1], H[i])
            )
        )
    return dp_table[-1]
```
