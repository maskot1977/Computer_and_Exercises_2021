```python
pokemon = {}
next_pokemon = {}
small_letters = {
    "ァ": "ア", "ィ":"イ", "ゥ":"ウ", "ェ":"エ", "ォ": "オ", 
    "ャ":"ヤ", "ュ":"ユ", "ョ":"ヨ"
}
for line in open("pokemon_names.txt"):
    name = line.strip()
    top = name[0]
    last = name[-1]
    if last == "ー":
        last = name[-2]
    if last in small_letters.keys():
        last = small_letters[last]
    
    # print(name, top, last)
    pokemon[name] = [top, last]
    if top not in next_pokemon.keys():
        next_pokemon[top] = []
    next_pokemon[top].append(name)
```

```python
import copy
import random
import time

patience = 20
pokemon_names = list(pokemon.keys())
random.shuffle(pokemon_names)
name = pokemon_names.pop()
waiting = [[name]]

longest = 0
start_time = time.time()
while len(waiting) > 0:
    current_time = time.time()
    if current_time - start_time > patience:
        start_time = time.time()
        if len(pokemon_names) > 0:
            name = pokemon_names.pop()
            waiting = [[name]]
        else:
            break
    names = waiting.pop()
    name = names[-1]
    if pokemon[name][1] in next_pokemon.keys():
        next_names = next_pokemon[pokemon[name][1]]
        random.shuffle(next_names)
        for next_name in next_names:
            new_names = copy.deepcopy(names)
            if next_name not in new_names:
                new_names.append(next_name)
                waiting.append(new_names)
    else:
        if longest < len(names):
            start_time = time.time()
            longest = len(names)
            print(len(names), names)
```
